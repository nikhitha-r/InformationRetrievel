####**Part A**

Provide a normalized vector $\hat{s}$ ∈ $\mathbf{R}^p$,such that:
\begin{align}
\hat{s} = argmax_{s s.t.\lVert{s}\rVert = 1} s^T\Sigma\Sigma^Ts
\end{align}

##### **Solution** :
 Consider,
 \begin{align} 
 s = [s_{1} \; \; s_{2} \; \;s_{3} \; \;s_{4} \; \;..\; s_{p}] ^ T
\end{align}
Then,
\begin{align}
s^T\Sigma\Sigma^Ts = \sigma_{1}^2s_{1}^2 + \sigma_{2}^2s_{2}^2 + \sigma_{3}^2s_{3}^2 + \:... \:+ \sigma_{p}^2s_{p}^2 = \Sigma_{i=1}^{p}\sigma_{i}^2s_{i}^2
\end{align}
Given *s* is normalised,

\begin{align}
\lVert{s}\rVert = 1
\end{align}
Therefore,
\begin{equation}
\lVert{s}\rVert^2 = 1  \;\;\;\;\;\;eqn. 1
\end{equation}
Since, $\sigma_{1}$ ≥ $\sigma_{2}$. . . ≥ $\sigma_{p}$ and from eqn. 1, when we maximize $s^T\Sigma\Sigma^Ts$:

\begin{align}
max_{\lVert{s}\rVert = 1} s^T\Sigma\Sigma^Ts = max_{\lVert{s}\rVert = 1} \Sigma_{i=1}^{p}\sigma_{i}^2s_{i}^2 = \sigma_{1}^2
\end{align}
Thus, to attain the maximum we have to set the coefficient of $\sigma_{1}^2$ to 1 and the others to 0:
\begin{align}
s_{1}^2 = 1, s_{i}^2 = 0 \; \forall \; i \neq 1
\end{align}
This implies that,
\begin{align}
\hat{s} \in \{ (1 \;\;0 \;\;0 \;..\; 0)^T , (-1 \;\;0\;\;0\;.. \; 0)^T\}
\end{align}

#### **Part A**
Write a function that takes as an input matrix T of vectorized images from subset 0. The output of this function are the 20 first singular vectors $\mathbf{U}$[:, 1],....,$\mathbf{U}$[:, 20].
Display the first 3 vectors as images, i.e., reshape them to size 50x50 and display
them.

```
import pandas as pd
import numpy as np
from PIL import Image
import os
import glob
from sklearn.decomposition import TruncatedSVD
import matplotlib.pyplot as plt
from sklearn.utils.extmath import randomized_svd
import re
from sklearn.neighbors import KNeighborsClassifier
from imageio import imread
import imageio
from sklearn.metrics import accuracy_score
```

```
"""
Fetch the images from the path and create an array
Input: datapath - directory path
"""
def import_images(datapath):
    imgs = os.listdir(datapath)
    img_array = []
    # Loop through the images and store in ann array
    for img in imgs:
        img_array += [np.array(imageio.imread(datapath+img)).astype(np.float64)]
    img_array = np.array(img_array)
    # Reshape the array. This will be of the size (70, 2500) for subset0
    img_array = np.reshape(img_array,(img_array.shape[0],img_array.shape[1]*img_array.shape[2]))
    return img_array

"""
Get the left singular vectors given the data and the number of vectors to be fetched.
Input: mat - data matrix
       num - number of left singular vectors
"""
def get_left_sing_vec(mat, num):
    # Centralize the matrix
    mu = np.mean(mat, axis = 1).reshape(mat.shape[0], 1)
    mat_c = mat - mu
    # Take SVD of mat_c
    U, Sigma, VT = np.linalg.svd(mat_c, full_matrices = True)
    # Return first num singular vectors
    return U.T[:num]
```

```
# Import the images in the folder subset0
data = import_images('yaleBfaces/subset0/')
print(data.shape)
# Get 20 left singular vectore
U = get_left_sing_vec(data.T, 20)
# Reshape
U_20 = np.reshape(U, (len(U), 50, 50))
print(U.shape)
```

```
# Plot first three left singular vectors
print("First 3 singular vectors")
fig, axes = plt.subplots(1, 3, figsize=(10,3))
for i in range(3):
    axes[i].imshow(U_20[i])
```

####**Part B**

Write a function that takes as an input the training set T (a matrix composed of
vectorized pictures from subset 0), a vector containing the labels of the training
set (i.e., if the the i-th sample belongs to class j, the i-th entry of the labels vector
is j. In this exercise j is an integer between 1 and 10), the test samples S (a
matrix composed of vectorized pictures from subsets 1-4) and the corresponding
labels (in a separate vector), the 20 singular vectors from the first step, and
the parameter k that denotes how many of the PCs are used. Use the Euclidean
distance to classify each sample image based on its three nearest neighbors. (This
is done by comparing the test samples with the training samples in the reduced
space.) As an output give the fraction of images from S that were misclassified,
 i.e., the error rate. Repeat this for subsets 1 through 4 and for k = 1,...., 20.
Plot the error rate for each subset

```
Train_s=data.T
Train_y=[]
files = os.listdir("yaleBfaces/subset0")
for file in files:
    Train_y+=[int(file.split("person")[1].split("_")[0])]
Train_y=np.array(Train_y)

k=20
#Get the first 20 left singular vectors of the Training set
U_20 = get_left_sing_vec(Train_s, k)

# Function to classify each sample image based on its three nearest neighbors based on Euclidean distance
def knn_classify(train_set, train_label, test_set, test_label, U_20, k):
    Reduced_Training = np.dot(train_set.T , U_20.T[:,:k])
    Reduced_test = np.dot(test_set.T , U_20.T[:,:k])
    
    estimate_labels = []
    
    for test in Reduced_test:
        #Calculate the euclidean distance of a single test image against all the training set
        distances = np.linalg.norm(Reduced_Training-test,axis=1)

        #Get the indices of the 3 closest ones
        min_distances_indices = np.argsort(distances)[:3]

        #Get the three shortest distances
        min_distances = distances[min_distances_indices.tolist()]

        #Change the minimum distance to prevent numerical instability
        x_min_distance = min_distances - np.mean(min_distances)

        #Get the labels of those three nearest neighbours
        estimated_labels = train_label[min_distances_indices]

        #Use weighted distances to get the estimated distance among the three nearest neighbours
        weights_sum = np.exp(-x_min_distance[0])+np.exp(-x_min_distance[1])+np.exp(-x_min_distance[2])
        estimate_label=0
        
        #Sum each neighbour label multiplied by its weighted distance
        for i in range(len(estimated_labels)):
            estimate_label += (np.exp(-x_min_distance[i])/weights_sum)*estimated_labels[i]
        estimate_labels+=[round(estimate_label)]
    
    #Calculate the accuracy of our results   
    test_acc = ( estimate_labels == test_label).mean()
    return 1-test_acc

Test_sets={}
Test_s=[]
Test_y=[]

# Get i error rates for each test set
for i in range(1,5):
    Test_sets["Test_set-"+str(i)]=[]
    
    #import the test set and the labels
    files = os.listdir("yaleBfaces/subset"+str(i)+"/")
    Test_s=import_images("yaleBfaces/subset"+str(i)+"/").T
    
    Test_y=[]
    for file in files:
        Test_y+=[int(file.split("person")[1].split("_")[0])] 
    Test_y=np.array(Test_y)
    
    #Varying the number of PCs used
    for j in range(1,k+1):
        Test_sets["Test_set-"+str(i)]+=[knn_classify(Train_s,Train_y,Test_s,Test_y,U_20,j)]

    #Add results to the plot    
    plt.plot(np.arange(1,k+1),Test_sets["Test_set-"+str(i)], '-o', label="Test_set-"+str(i))

plt.plot(np.arange(1,k+1),np.full(k, 0.9), '-o', label="Random guessing")
    
plt.title('Error rate')
plt.xlabel('Singular Vectors Used :K')
plt.ylabel('Error Rate')
plt.legend(loc='lower right')
plt.gcf().set_size_inches(16, 12)
plt.show()
```

**Observations**
* There is a big difference among the test subsets, and a big improvement in accuracy when we drop the first 3 principal components.

* The images in the subsets progressively becomes less illuminated .i.e Subset1 has much better illumination compared to subset 4. The details are captured better when the illumination is better and hence are better classified.

* The improvement in accuracy can be explained by looking at the first 3 principal component vectors. When plotted as images, they look like negatives of the images, capturing brightness information of the set, which are useless features for the facial clasification task. With brightness out of the picture, the principal vectors focus more on the facial features giving higher accuracy. The exception for this rule is subset 4 since the brightness is so low that it is difficult to capture the details of the facial features. Hence, losing brighness hurts the accuracy for this subset.

* It can be obserevd that for subset 1, the removal of first three cols of $U_k$ does not particularly affect the performance. This may be because the first three principal components are as good at separating samples by labels compared to other principal components.

* As opposed to subset 1, we can observe better performance for subsets 2 and 3 on removing the first three cols of $U_k$. This may be because even though the first three components capture more variance in the samples, may be the other components are better at separating samples by labels, allowing KNN to correctly classify samples.

* For subset 4, a worse performance is observed on removing first three cols of $U_k$. This may be because the first three principal components are better at separating samples by labels compared to the other principal components. It should be noted that both errors remained quite high (around $0.9$), so the strength of this analysis is not as strong as the analysis of the other subsets.
